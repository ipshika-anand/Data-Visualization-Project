**1] FOR GENDER**

```python
import plotly.graph_objects as go

labels = ['Female', 'Male']
sizes = [Gender['Female'].sum(), Gender['Male'].sum()]
colors = ['tan', 'lightcoral']

fig = go.Figure(data=[go.Pie(labels=labels, values=sizes, marker=dict(colors=colors), textinfo='percent',
                             hoverinfo='label+percent+value')])

fig.update_layout(title='Male vs Female Athletes')

fig.show()
```

```python
import plotly.express as px

px.bar(Gender, x="Discipline", y="Total", color="Discipline",
       title="Total participants in each Discipline : ")
```

```python
#BAR CHART
import plotly.express as px

# Bar chart with both male and female participants
bar_chart = px.bar(Gender, x="Discipline", y=["Male", "Female"],
                   title="Male and Female Participants in Each Discipline",
                   labels={"value": "Total participants", "variable": "Gender"},
                   barmode='group', color_discrete_map={"Male": "blue", "Female": "orange"})

# Show bar chart
bar_chart.show()
```

```python
#STACKED BAR CHART
import plotly.express as px

# Stacked bar chart
stacked_bar_chart = px.bar(Gender, x="Discipline", y=["Male", "Female"],
                           title="Male and Female Participants in Each Discipline",
                           labels={"value": "Total participants", "variable": "Gender"},
                           color_discrete_map={"Male": "blue", "Female": "orange"},
                           barmode='stack')

# Show stacked bar chart
stacked_bar_chart.show()
```

```python
sunburst = px.sunburst(Gender, path=['Discipline'], values='Total')
sunburst.show()
```

```python
px.scatter(Medals1, x="Total", y="Rank by Total",size="Total", color="Team/NOC",hover_name="Team/NOC", log_x=True, size_max=60)
```

```python
medals = pd.read_excel('Medals.xlsx')
color_scale_reversed = "blues_r"

# Create the choropleth map
fig = px.choropleth(medals,
                    locations="Team/NOC",
                    locationmode='country names',
                    color="Rank by Total",
                    hover_name="Total",
                    range_color=[1, 100],
                    color_continuous_scale=color_scale_reversed)

fig.show()
```
